## Gender validation using regular expresions

Read gender inputs from ``genders.txt`` file. 

Valid formats are ``female``, ``Female``, ``male``, or ``Male``. 

Anything other is invalid.

```python
# read genders from file and remove new lines
genders = [gender.rstrip() for gender in open('genders.txt')]

print(genders[:10], len(genders))
```

```python
def validate_genders(genders, N, regex):
    
    N = N if N<=len(genders) else len(genders)
    print(N)
    for gender in genders[:N+1]:
        res = regex.fullmatch(gender)
        if res == None:
            print(gender,'is invalid')
        else:
            print(gender,'is valid')
```

Lets find ``female``, ``Female``, ``male``, or ``Male``

```python
import re

regex = re.compile('(male|female)',re.I)

validate_genders(genders, 100, regex)
```

```python
regex = re.compile('(fe){0,1}male',re.I)

validate_genders(genders, 100, regex)
```

Lets make any capital letters (except first) invalid.

For example ``Female`` is valid, ``femaLe`` is invalid.

```python
regex = re.compile('[Ff]emale|[Mm]ale')
#regex = re.compile('[Mm]ale')

validate_genders(genders, 100, regex)
```
