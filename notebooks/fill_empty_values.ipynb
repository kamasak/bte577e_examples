## Fill empty values using pandas library

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, np.nan],
                   [np.nan, 3, np.nan, 4]],
                  columns=list("ABCD"))

df
```

```python
ind = pd.isna(df)

ind
```

```python
df1 = df.copy()
df1[ind] = -7

df1
```

Replace all NaN elements with 0s. If inplace argument is not True (default is False) dataframe is not changed, a new data frame is returned.

```python
df.fillna(0, inplace=False)

df
```

```python
df2 = df.copy()
df2.fillna(0, inplace=True)

df2
```

```python
df3 = df.fillna(0)

df3
```

propagate non-null values forward or backward

``ffill``: forward fill
``bfill``: backward fill

```python
df1 = df.fillna(method="bfill", axis=0)

df
```

```python
df1
```

```python
df1 = df.fillna(method="bfill", axis=1)

df1
```

```python
df1 = df.fillna(method="ffill", axis=0)

df1
```

```python
df1 = df.fillna(method="ffill", axis=1)

df1
```

```python
df1 = df.fillna(method="ffill", axis=0, limit=1)

df1
```
